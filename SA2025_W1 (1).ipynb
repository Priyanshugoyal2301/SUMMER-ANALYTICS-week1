### 1. What is name of car that has the highest horsepower?

### 2. How many cars have mpg ≥ 35?

```python
count_high_mpg = data[data['mpg'] >= 35].shape[0]
print("Number of cars with mpg ≥ 35:", count_high_mpg)
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
filtered_cars = data[(data['horsepower'] > 100) & (data['weight'] < 3000)]

most_common_origin = filtered_cars['origin'].mode()[0]
print("Most common origin for cars with horsepower > 100 and weight < 3000:", most_common_origin)
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
japan_cars = data[data['origin'] == 'japan']

mean_acceleration_japan = round(japan_cars['acceleration'].mean(), 2)
print("Mean acceleration of cars from Japan:", mean_acceleration_japan)
```

### 5. Which year had the highest average mpg?

```python
avg_mpg_per_year = data.groupby('model_year')['mpg'].mean()

year_highest_mpg = avg_mpg_per_year.idxmax()
print("Year with the highest average mpg:", year_highest_mpg)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
median_mpg = data['mpg'].median()

above_median_mpg_cars = data[data['mpg'] > median_mpg]

above_median_mpg_cars['hp_per_weight'] = above_median_mpg_cars['horsepower'] / above_median_mpg_cars['weight']

max_ratio = above_median_mpg_cars['hp_per_weight'].max()

best_cars = above_median_mpg_cars[above_median_mpg_cars['hp_per_weight'] == max_ratio]

print("Car(s) with the best horsepower-to-weight ratio among cars with above-median mpg:")
print(best_cars)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
import seaborn as sns
import matplotlib.pyplot as plt

avg_mpg_year_origin = data.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.lineplot(data=avg_mpg_year_origin, x='model_year', y='mpg', hue='origin', marker='o')

plt.xlabel('Year')
plt.ylabel('Average MPG')
plt.title('Average MPG over Years by Origin')
plt.legend(title='Origin')
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
name_counts = data['name'].value_counts()

multi_year_names = name_counts[name_counts > 1].index

filtered_data = data[data['name'].isin(multi_year_names)]

grouped = filtered_data.groupby('name').agg(
    appearances=('model_year', 'nunique'),
    mpg_mean=('mpg', 'mean'),
    mpg_std=('mpg', 'std')
).reset_index()

consistent_candidates = grouped[(grouped['appearances'] >= 2) & (grouped['mpg_std'] < 1)]

min_std = consistent_candidates['mpg_std'].min()

best_consistent = consistent_candidates[consistent_candidates['mpg_std'] == min_std]

final_sorted = best_consistent.sort_values(by=['appearances', 'mpg_mean'], ascending=[False, False])

report = final_sorted[['name', 'appearances', 'mpg_mean']]
print(report)
```
